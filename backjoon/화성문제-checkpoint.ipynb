```python
T = int(input())

## 반복횟수(T)
for i in range(T):
## num 리스트 초기화
    num = []
    num = input().split()
    cnt = float(num[0]) ## cnt 초기화
    del num[0]
    
## 계산 순서대로 계산하기 위해    
    for j in num:
        if "@" in j:
            cnt = cnt*3   
        if "%" in j:
            cnt = cnt+5           
        if "#" in j:
            cnt = cnt-7            
    print("{:.2f}".format(cnt))
    
## 고려하지 못한 사항 계산순서
```
